```python
# import librearies.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas.plotting import scatter_matrix
%matplotlib notebook
```

```python
# Loading cleaned data.
# df = training data
# df2 = testing data

df = pd.read_csv('cleaned_training.csv', index_col='idx')
df2 = pd.read_csv('cleaned_testing.csv', index_col='idx')
```

```python
# Getting the correlations.
corr_matrix = df.corr()

corr_matrix['RUL'].sort_values(ascending=False)
```

```python
# plotting the scatter matrix of correlated point
attribs = ['RUL', 'time_cycles', 'sem_12', 'sem_7', 'sem_9', 'sem_14', 'sem_8', 'sem_6', 'sem_15']
scatter_matrix(df[attribs], figsize=(12, 8))
```
